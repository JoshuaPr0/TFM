```python
import pandas as pd
import numpy as np

df = pd.read_csv('tweets_limpios_sin_HT.csv', sep=',')
```

```python
n = df.nunique(axis=0)
n
```

```python
df2 = df[["Usuario", "Followers"]]

masSeguidos = df2[['Usuario', "Followers"]].drop_duplicates().nlargest(435, ['Followers'])

masSeguidos = masSeguidos['Usuario'].drop_duplicates()

masSeguidos = masSeguidos.reset_index(drop=True)

for i in masSeguidos:
  print(i)
```

```python
listado = []

for i in masSeguidos:
  for j in df['Usuario']:
    if i == j:
      listado.append(j)

listado2 = list(set(listado))
df3 = df[0:0]

for i in listado2:
  filas = df.loc[df['Usuario'] == i]
  df3 = pd.concat([filas, df3], axis=0)

df3 = df3.drop_duplicates().reset_index(drop=True)

df3
```

```python
df3.to_csv('Tuits_100_Usuarios_Followers.csv', sep=',', index=False)
```

```python
df["Rango Retweets"] = df.groupby('Usuario')['Retweets'].transform(np.sum).rank(ascending=True, method='dense')
df["Rango Likes"] = df.groupby('Usuario')['Likes'].transform(np.sum).rank(ascending=True, method='dense')
df["Rango Respuestas"] = df.groupby('Usuario')['Respuestas'].transform(np.sum).rank(ascending=True, method='dense')
df["Rango Citados"] = df.groupby('Usuario')['Citados'].transform(np.sum).rank(ascending=True, method='dense')
df["Rango Vistas"] = df.groupby('Usuario')['Vistas'].transform(np.sum).rank(ascending=True, method='dense')
df["Rango Total"] = df["Rango Retweets"] + df["Rango Likes"] + df["Rango Respuestas"] + df["Rango Citados"] + df["Rango Vistas"]
```
