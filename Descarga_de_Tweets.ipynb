```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import csv
import warnings
warnings.filterwarnings("ignore")

tweets = []
```

```python
maxTweets = 500000
for i,tweet in enumerate(sntwitter.TwitterSearchScraper(query = "#diainternacionaldelamujer since:2023-03-05 until:2023-03-10").get_items()):
    if i>maxTweets:
        break
    tweets.append([tweet.hashtags, tweet.date, tweet.rawContent, tweet.user.username, tweet.user.followersCount, tweet.replyCount, tweet.retweetCount, tweet.likeCount, 
                   tweet.quoteCount, tweet.viewCount, tweet.retweetedTweet, tweet.lang]) 

df = pd.DataFrame(tweets, columns=['Hashtags', 'Fecha', 'Texto', 'Usuario', 'Followers', 'Respuestas', 'Retweets', 'Likes', 'Citados', 'Vistas', 'Es_Rewteet', 'Idioma'])

df.to_csv('out-tweets.csv', sep=',', index=False)
```
