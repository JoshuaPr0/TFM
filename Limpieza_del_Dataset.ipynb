```python
import pandas as pd

df1 = pd.read_csv('out-tweets diainternacionaldelamujer.csv', sep=',')
df2 = pd.read_csv('out-tweets-2 8m2023.csv', sep=',')
df3 = pd.read_csv('out-tweets-3 DiaDeLaMujer.csv', sep=',')
df4 = pd.read_csv('out-tweets-4 8DeMarzo.csv', sep=',')

df_tweets = pd.concat([df1, df2, df3, df4])

print("\nHashtag #diainternacionaldelamujer:")
print(df1.shape)
print("\nHashtag #8m2023:")
print(df2.shape)
print("\nHashtag #DiaDeLaMujer:")
print(df3.shape)
print("\nHashtag #8DeMarzo:")
print(df4.shape)
print("\nTotal de los datos extraídos:")
print(df_tweets.shape)
df_tweets.head(5)
```

```python
df_tweets.info()
print('\n-----------------------\n')
print('Recuento de nulos')
df_tweets.isnull().sum()
```

```python
df_tweets_limpieza = df_tweets

# Reemplazamos los nulos de Vistas por 0s
df_tweets_limpieza['Vistas'].fillna(0, inplace = True)

# Eliminamos la columna vacía
df_tweets_limpieza = df_tweets_limpieza.drop('Es_Rewteet', axis=1)

df_tweets_limpieza.head(5)
```

```python
# Revisamos los idiomas que aparecen
gb = df_tweets_limpieza.groupby(['Idioma'])['Idioma'].count()
gb
```

```python
# qht son hashtags y qme son hashtags y arrobas.
# Nos quedamos solo con los tweets que tengan lenguaje Español para facilitar el análisis de sentimientos
df_tweets_limpieza = df_tweets_limpieza[df_tweets_limpieza['Idioma'] == 'es']

# Eliminamos posibles duplicados al descargar tweets por hashtag
df_tweets_limpieza = df_tweets_limpieza.drop_duplicates(('Texto', 'Usuario', 'Fecha'))

df_tweets_limpieza.shape
```

```python
df_tweets_limpieza = df_tweets_limpieza[df_tweets_limpieza['Hashtags'].notnull()]

df_tweets_limpieza.shape
```

```python
df_tweets_limpieza.to_csv('tweets_limpios_sin_HT.csv', sep=',', index=False)
```

```python
from collections import Counter

total = df_tweets_limpieza['Texto']

listado = []

def extract_hashtags(text):
 
    # initializing hashtag_list variable
    hashtag_list = []
 
    # splitting the text into words
    for word in text.split():
 
        # checking the first character of every word
        if word[0] == '#':
 
            # adding the word to the hashtag_list
            hashtag_list.append(word[1:])
 
    for hashtag in hashtag_list:
        listado.append(hashtag)

for w in total:
  extract_hashtags(w)

d = Counter(listado)
d.most_common(20)
```
